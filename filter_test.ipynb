```python
import scripts.emg_transform as emg_transform
import numpy as np
import matplotlib.pyplot as plt


# Sample EMG data (replace with your actual data)
fs = 10000  # Sampling frequency (Hz)
t = np.linspace(0, 0.5, fs * 5, False)  # 5 seconds of data
emg_data = np.sin(2 * np.pi * 60 * t) + np.random.normal(0, 2, len(t))  # Simulated EMG data with 60Hz noise

```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, emg_data, color='blue')
plt.title('EMG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Apply the zero-phase bandpass filter
filtered_data = emg_transform.butter_bandpass_filter(emg_data, fs)  # Pass frequencies between 20Hz and 200Hz
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, filtered_data, color='blue')
plt.title('EMG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
